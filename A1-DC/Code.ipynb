It has dropped a lot of rows, so we will try to minimize the rows that we will drop by removing unnecessary columns and replacing some values

```python
#Drop the coulmns
data_no_un_coulmns = data.drop(columns=['show_id', 'cast', 'director', 'description'])
#Replacments
data_no_un_coulmns['country'] = data['country'].fillna(data['country'].mode()[0])
```

```python
#
data_no_un_coulmns_noEmpty = data_no_un_coulmns.dropna()
print(f'We have drop {data.shape[0]-data_no_un_coulmns_noEmpty.shape[0]} empty rows')
data_no_un_coulmns_noEmpty.drop_duplicates()
print(f'We have drop {data.shape[0]-data_no_un_coulmns_noEmpty.shape[0]} empty rows')
```

as you say , now we just dropped 17 raws 

```python
ratings_ages = {
    'TV-PG': '+13 years',
    'TV-MA': 'Adults',
    'TV-Y7-FV': '+13 years',
    'TV-Y7': '+13 years',
    'TV-14': '+17 years',
    'R': '+20 years',
    'TV-Y': '+3 years',
    'NR': '+3 years',
    'PG-13': '+17 years',
    'TV-G': '+3 years',
    'PG': '+13 years',
    'G': '+3 years',
    'UR': 'Adults',
    'NC-17': 'Adults'
}

# Replace Rating values with age targets
# data_no_un_coulmns_noEmpty['ages'] = data_no_un_coulmns_noEmpty['rating'].replace(ratings_ages)
# data_no_un_coulmns_noEmpty['ages'].unique()
# data_no_un_coulmns_noEmpty.head(10)

data_no_un_coulmns_noEmpty.loc[:, 'ages'] = data_no_un_coulmns_noEmpty['rating'].replace(ratings_ages)
unique_ages = data_no_un_coulmns_noEmpty['ages'].unique()
print("Unique ages:", unique_ages)
data_no_un_coulmns_noEmpty.head(10)
```

```python
data_no_un_coulmns_noEmpty.to_csv('netflix_titles_new.csv',index=True)
```
